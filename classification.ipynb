```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(drop='first',sparse=False)
```

```python
df['International plan'].value_counts()
```

```python
df1 = encoder.fit_transform(df[['International plan','Voice mail plan']])
```

```python
df1 = pd.DataFrame(df1,columns=['International_plan','Voicemail_plan'])
```

```python
df1
```

```python
df2 = pd.concat([df,df1],axis=1)
```

```python
df2
```

```python
df3 = encoder.fit_transform(df[['Churn']])
```

```python
df3 = pd.DataFrame(df3,columns=['churn'])
```

```python
df3
```

```python
df_final = pd.concat([df2,df3],axis=1)
```

```python
df_final
```

```python
df_final = df_final.drop(['International plan','Voice mail plan','Churn'],axis=1)
```

```python
df_final
```

```python
X = df_final.drop(['churn','State'],axis=1)
```

```python
y = df_final['churn']
```

```python
X.head()
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
model1 = RandomForestClassifier()
```

```python
parameters = {'n_estimators':[34,65,10,500,400,250,300],'criterion':['gini','entropy']}
```

```python
grid_cv = GridSearchCV(model1,param_grid=parameters,cv=5)
```

```python
grid_cv.fit(X_train,y_train)
```

```python
grid_cv.best_params_
```

```python
model2 = RandomForestClassifier(criterion='gini',n_estimators=65)
```

```python
model2.fit(X_train,y_train)
```

```python
y_pred2 = model2.predict(X_test)
```

```python
print(classification_report(y_test,y_pred2))
```

## Drastic improvement from 91% to 92%
